```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')          #internal dataset
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
#IN OUR DATA WE HAVE EVENT WHICH ARE TWO-STIM AND CUE , SO WE MAKE SEPARATE LINES FOR THESE 2 EVENTS
#SO MAP THAT COLLUMN TO HUE PARAMETER, HUE DETERMINES LINE COLOUR - SO NOW EVENT DETERMINES IT

sns.lineplot(x='timepoint',y='signal',data=fmri, hue='event')
plt.show()
```

```python
#style determines style of line so now it will be represented by evemt

sns.lineplot(x='timepoint',y='signal',data=fmri, hue='event',style='event')
plt.show()
```

```python
# markers is used tell exact signal at exact time point
sns.lineplot(x='timepoint',y='signal',data=fmri, hue='event',style='event',markers=True)
plt.show()
```

# Barplot

```python
import pandas as pd
sns.set(style='whitegrid')       #grid method
pokemon=pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.barplot(x='Legendary',y='Speed',data=pokemon)               #o or 1 legendary
plt.show() 
```

```python
sns.barplot(x='Legendary',y='Total',data=pokemon)               #o or 1 legendary
plt.show() 
```

```python
sns.barplot(x='Legendary',y='Speed',hue='Generation',data=pokemon)               #o or 1 legendary
plt.show() 
```

```python
sns.barplot(x='Legendary',y='Speed',palette="Blues_d",data=pokemon) #palette is basically a colour it can be rocket, vlag, etc              #o or 1 legendary
plt.show() 
```

```python
#use color for fixed color or for bars
sns.barplot(x='Legendary',y='Speed',color='green',data=pokemon)               #o or 1 legendary
plt.show() 
```

# Histogram

```python
# here we dont get hg directly we get distrubtuion plot it contains histogram and a freq curve also
pop=pd.read_csv('countries-table.csv')

sns.distplot(pop['worldPercentage'])   
plt.show()
```

```python
sns.distplot(pop['worldPercentage'],hist=False,color='purple')    #histogram nt shown
plt.show()
```

```python
sns.distplot(pop['worldPercentage'],kde=False,color='purple',bins=5)    #using kde false remove curve
plt.show()
```

```python
sns.distplot(pop['worldPercentage'],color='purple',vertical=True)  ## now vertically in y axis distributn on x asis
```

# Box plot

```python
# used between a char/categorical(x axis) and numerical var (y axis)   
import pandas as pd
poke=pd.read_csv("pokemon.csv")
sns.boxplot(x='Legendary',y='Speed',data=poke)          #IMP  # WE USE MEDIAN VALU FOR COMPARISTION
plt.show()                                              # THICK OUTLIRS
```

```python
sns.boxplot(x='Generation',y='Attack',data=poke,palette='Set1')  #CAN also use colour
plt.show()
```

```python
# using linewidth increase width o fboundry line
sns.boxplot(x='Legendary',y='Speed',data=poke,linewidth=3)
plt.show()
```

```python
# we can also change the order acc to us
sns.boxplot(x='Generation',y='Attack',data=poke,palette='Set1',order=[6,3,2,5,1,4])  #use order, we can decide by ourselve what to put first
plt.show()
```

```python
sns.boxplot(x='Generation',y='Attack',data=poke,palette='Set1',hue='Legendary')
plt.show()
```
